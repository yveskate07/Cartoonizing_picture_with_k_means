```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Configuration pour affichage dans le notebook
plt.rcParams['figure.figsize'] = [10, 8]
```

Une image couleur est composée de 3 canaux :

R = Rouge (Red)

G = Vert (Green)

B = Bleu (Blue)

Chaque pixel est donc un triplet : (R,G,B)

RGB est l’ordre standard utilisé par :

Matplotlib

PIL

la majorité des librairies d’affichage

OpenCV, pour des raisons historiques et de performance, stocke les couleurs dans l’ordre (B,G,R)
Raisons principales :

Héritage historique (anciennes APIs C/C++)

Compatibilité avec certains formats bitmap

Optimisations bas niveau

```python
def show_image(title, image): # image est un tableau NumPy (numpy.ndarray).
    """Affiche une image dans le notebook avec matplotlib"""
    # OpenCV utilise BGR, Matplotlib utilise RGB
    if len(image.shape) == 3:  # Image couleur
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    else:  # Image en niveaux de gris
        image_rgb = image
    
    plt.figure(figsize=(10, 8))
    plt.imshow(image_rgb, cmap='gray' if len(image.shape) == 2 else None)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def read_file(filename):
  img = cv2.imread(filename)
  cv2.imshow(img)
  return img
```

data → pixels sous forme (N, 3)

k → nombre de couleurs finales

None → pas de labels initiaux

criteria → quand s’arrêter

10 → nombre de relances

KMEANS_RANDOM_CENTERS → centres initiaux aléatoires

Réduit le nombre de couleurs d’une image en appliquant une quantification de couleurs
par K-means.
On remplace chaque pixel par la couleur moyenne du groupe auquel il appartient.
L’image est transformée en un ensemble de pixels vus comme des points dans l’espace
couleur (B, G, R). L’algorithme K-means regroupe ces pixels en k clusters selon leur
similarité de couleur, puis chaque pixel est remplacé par la couleur moyenne de son
cluster.

Le résultat est une image avec k couleurs dominantes, produisant des aplats de couleur
caractéristiques d’un effet de type "cartoon".
Pourquoi on parle de "quantification" ?

K-Means remplace :

un continuum (toutes les couleurs possibles)

par un ensemble fini (k couleurs)

```python
def color_quantization(img, k):
# Transform the image
  data = np.float32(img).reshape((-1, 3))

# Determine criteria
  criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 20, 0.001)

# Implementing K-Means
  ret, label, center = cv2.kmeans(data, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
  center = np.uint8(center)
  result = center[label.flatten()]
  result = result.reshape(img.shape)
  return result
```

"""
Génère un masque de contours à partir d’une image couleur.

L’image est d’abord convertie en niveaux de gris puis lissée à l’aide d’un flou médian
afin de réduire le bruit tout en préservant les bords. Un seuillage adaptatif est ensuite
appliqué pour détecter les transitions d’intensité et produire une image binaire où les
contours apparaissent sous forme de lignes noires.

Ce masque de contours est utilisé pour renforcer l’effet de type "cartoon" en ajoutant
des lignes nettes autour des formes.
"""


```python
def edge_mask(img, line_size, blur_value):
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  gray_blur = cv2.medianBlur(gray, blur_value)
  edges = cv2.adaptiveThreshold(gray_blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, line_size, blur_value)
  return edges
```

**Run Code to Create Cartoon Effect**

```python
line_size = 7
blur_value = 7

edges = edge_mask(img, line_size, blur_value)
show_image('Affichage des contours', edges)
```

```python
total_color = 4
img = color_quantization(img, total_color)
show_image('Quantification des couleurs', img)
```

```python
blurred = cv2.bilateralFilter(img, d=7, sigmaColor=200, sigmaSpace=200)
show_image('Image floutée (bilateral filter)', blurred)
```

Cette cellule :

applique un flou bilatéral à l’image quantifiée

lisse les zones de couleur sans toucher aux contours

améliore la propreté visuelle du rendu

prépare l’image à être combinée avec le masque de contours

```python
cartoon = cv2.bitwise_and(blurred, blurred, mask=edges)
show_image('Résultat Cartoon', cartoon)
```
